### PCM 

```python
A1 = np.random.randint(0,4)
A2 = np.random.randint(0,4)
t = np.linspace(0,5,1000)

f1 = np.random.randint(0,5)
f2 = np.random.randint(0,5)

x = A1*(1+np.sin(2*np.pi*f1*t)) + A2*(1+np.sin(2*np.pi*f2*t))

plt.plot(t,x)

s_rate = 10
T = 1/s_rate

n = np.linspace(0,5/T)
nT = n*T

t = np.linspace(0,5,5*s_rate)
S = A1*(1+np.sin(2*np.pi*f1*nT)) + A2*(1+np.sin(2*np.pi*f2*nT))
plt.plot(t,S, "ro")
```

```python
q = [0,2,4,6,8,10,12,14,16]
Q = []
for i in S:
    for j in range(0,9):
        if(i>=q[j] and i<=q[j+1]):
            Q.append(int((q[j]+q[j+1])/2))
            
print('Q = ',Q)        
```

### BPSK

```python
#carrier
fc=100
A=1
w=100
t=np.linspace(0,6*np.pi,w)
carrier=A*np.cos(2*np.pi*fc*t)

magnitude_carrier_square=[(np.abs(carrier))**2] #sum(|x|**2)=Energy
E=np.sum(magnitude_carrier_square)
carrier_normalised=carrier/(np.sqrt(E))
plt.plot(carrier_normalised);
```

```python
n = 10

#------------ Transmitter---------------
inputSyms=np.round(np.random.rand(n))

s=2*inputSyms-1 

product=[i*carrier_normalised for i in s] #multipying with carrier
bpsk_seq=[j for k in range(n) for j in product[k]] #all the points in product to a single list

fig, ax = plt.subplots(nrows=1,ncols = 1)
fig.set_figwidth(20) #resizing the graph
ax.plot(bpsk_seq,"g");
ax.set_title("BPSK Waveform");
# plt.figure()

fig, ax = plt.subplots(nrows=1,ncols = 1)
plt.plot(np.real(s),np.imag(s),'r*');
ax.set_title("Constellation diagram of BPSK");
```

```python
#channel

berr=[]
snr = np.arange(start=-4,stop = 13, step = 2)
No=1/(10**(snr/10))
sigma=np.sqrt(No/2) # Finding Noise power
#print(sigma)

for i in range(len(sigma)):
    noise=sigma[i]*np.random.randn(1,n)
    rmsg =s+noise;# Adding noise through channel
    
    #-------------- Receiver ------------
    detectedSyms = (rmsg >= 0).astype(int) #thresolding at value 0
    #print(detectedSyms)
    berr.append((np.sum(detectedSyms != inputSyms)/n)) #calculate BER

    
#print(berr)

from scipy.special import erfc #erfc/Q function
BER_theory = 0.5*erfc(np.sqrt(10**(snr/10)))

fig, ax = plt.subplots(nrows=1,ncols = 1)
ax.semilogy(snr,berr,color='r',marker='o',linestyle='',label='BPSK Sim')
ax.semilogy(snr,BER_theory,marker='',linestyle='-',label='BPSK Theory')
ax.set_xlabel('$E_b/N_0(dB)$');ax.set_ylabel('BER ($P_b$)')
ax.set_title('Probability of Bit Error for BPSK over AWGN channel')
ax.grid()
ax.legend();plt.show()
```

### QPSK

```python
n = 10
x = np.random.randint(0,2,n)

even = x[0::2]
odd = x[1::2]

e_bits = 2*even-1
o_bits = 2*odd-1


f = 100
A = 1
w = 100
t = np.linspace(0,6*np.pi, w)
carrier = [A*np.cos(2*np.pi*f*t), A*np.sin(2*np.pi*f*t)]

prod_even = []
prod_odd = []
for i in range(n//2):
    prod_even.append(e_bits[i]*carrier[0])
    prod_odd.append(o_bits[i]*carrier[1])
```

```python
qpsk_even = [j for k in range(n//2) for j in prod_even[k]]
qpsk_odd = [j for k in range(n//2) for j in prod_odd[k]]
```

```python
qpsk = [qpsk_odd[i]+qpsk_even[i] for i in range(len(qpsk_even))]
```

```python
plt.plot(qpsk)
```

```python
mod=e_bits+1j*o_bits
#print(mod)
```

```python
berr=[]

snr = np.arange(start=-4,stop = 13, step = 2)
No=1/(10**(snr/10))
sigma=np.sqrt(No/2) # Finding Noise power
#print(sigma)

for i in range(len(sigma)):
    noise=sigma[i]*np.random.randn(1,n//2)
    rmsg =mod+noise[0];# Adding noise through channel
#     print(rmsg)
    
    #-------------- Receiver ------------
    
    detectedSyms_odd = (np.imag(rmsg) >= 0).astype(int) #thresolding at value 0
#     print(detectedSyms_odd)
    berr1=np.sum(detectedSyms_odd != odd)/(n//2) #calculate BER1
    
    detectedSyms_even = (np.real(rmsg) >= 0).astype(int) #thresolding at value 0
    berr2=np.sum(detectedSyms_even != even)/(n//2) #calculate BER2
    
    berr.append((berr1+berr2)/2)

    
#print(berr)

from scipy.special import erfc #erfc/Q function
BER_theory = 0.25*erfc(np.sqrt(10**(snr/10)))

plt.semilogy(snr,berr,"r*",snr,BER_theory,"b");
plt.title('Probability of Bit Error for QPSK over AWGN channel');
plt.xlabel("SNR");
plt.ylabel("BER");
```
